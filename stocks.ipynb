```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("all_stocks_5yr.csv")
df
```

```python
print(df.head())
```

```python
#Time Series Analysis of all stocks

df["date"] = pd.to_datetime(df["date"])
plt.plot(df["date"], df["close"])
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.title("S&P 500 Stock Price Over Time")
plt.show()
```

```python
#Filter only the FAANG companies
selected_stocks = ["GOOGL","AAPL","MSFT","AMZN","FB","NFLX"]

data_selected = df[df["Name"].isin(selected_stocks)]
```

```python
# Plot time series for the FAANG stocks
for stock_name, stock_data in data_selected.groupby('Name'):
    plt.plot(stock_data['date'], stock_data['close'], label=stock_name)

plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Stock Price Over Time - FAANG Stocks')
plt.legend()
plt.show()
```
